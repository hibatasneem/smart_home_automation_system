```python

#**************DEVICE MANAGEMENT ******************
def deviceManagement():
    homeAppliance = ["fan", "AC", "microwave", "door lock", "lights"]
    print(
        "Device management options:"
        "\n1. Add new device"
        "\n2. Remove device"
        "\n3. Sort devices alphabetically"
        "\n4. View devices"
        "\n5. Exit"
    )

    while True:
        choice = int(input("\nEnter your choice (1-5): "))

        if choice == 1:
            newDevice = input("Enter the new device you want to add: ")
            homeAppliance.append(newDevice)
            print(f"\nUpdated Devices: {homeAppliance}")

        elif choice == 2:
            removeDevice = input("Enter the device you want to remove: ")
            if removeDevice in homeAppliance:
                homeAppliance.remove(removeDevice)
                print(f"\nUpdated Devices: {homeAppliance}")
            else:
                print("\nDevice not found.")

        elif choice == 3:
            homeAppliance.sort()
            print(f"\nSorted Devices: {homeAppliance}")

        elif choice == 4:
            print(f"\nDevices: {homeAppliance}")

        elif choice == 5:
            print("\nExiting Device Management...")
            break

        else:
            print("\nInvalid choice. Please enter a number between 1 and 5.")

#********************ENERGY CONSUMPTION*******************
def energyConsumption():
    deviceEnergy = {
        "Fan": 0.031,
        "AC": 1.5,
        "Lights": 0.06,
        "Microwave": 1.2,
        "Door Lock": 0.0005
    }
    print(deviceEnergy)
    print("ENERGY CONSUMPTION TRACKER:"
          "\n1. Add new energy usage data"
          "\n2. Remove device energy usage"
          "\n3. Update existing usage"
          "\n4. Display total energy consumption"
          "\n5. Exit")

    while True:
        choice = int(input("Enter your choice (1-5): "))

        if choice == 1:
            key = input("Enter new device data (key): ")
            value = float(input("Enter new device data (value in kWh): "))
            deviceEnergy[key] = value
            print(f"After adding new device: {deviceEnergy}")

        elif choice == 2:
            remove = input("Enter the device you want to remove: ")
            if remove in deviceEnergy:
                deviceEnergy.pop(remove)
                print(f"Updated Energy Usage: {deviceEnergy}")
            else:
                print("Device not found.")

        elif choice == 3:
            key = input("Enter the device you want to update: ")
            if key in deviceEnergy:
                value = float(input("Enter the new energy usage (value in kWh): "))
                deviceEnergy[key] = value
                print(f"Updated Energy Usage: {deviceEnergy}")
            else:
                print("Device not found.")

        elif choice == 4:
            totalEnergy = sum(deviceEnergy.values())
            print(f"Total Energy Consumption: {totalEnergy} kWh")

        elif choice == 5:
            print("Exiting Energy Consumption Tracker...")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

#********************SMART HOME AUTOMATION **********************
def smartHomeAutomation():
    def calculateEnergyCost(energyUsage, ratePerKwh):
        totMonthlyCost = energyUsage * ratePerKwh
        print(f"Total monthly energy cost: $ {totMonthlyCost:.2f}")
    calculateEnergyCost(120,15)
    home1Device = input("Enter devices in your home 1 (separated by commas): ")
    home1Device = [device.strip() for device in home1Device.split(",")]
    print(f"Home 1 Devices: {home1Device}")

    home2Device = input("Enter devices in your home 2 (separated by commas): ")
    home2Device = [device.strip() for device in home2Device.split(",")]
    print(f"Home 2 Devices: {home2Device}")

    def findCommonDevice(home1Device, home2Device):
        return set(home1Device).intersection(home2Device)

    common = findCommonDevice(home1Device, home2Device)
    print("Common Devices:", common)



#**************SET AUTOMATION RULE *******************



def setAutomationRule():
    def automationRule(device, time, action):
        print(f"Automation Rule: {device} will be turned {action} at {time} .")
    automationRule("AC","11:00 PM","OFF")
    device = [("lights", 90), ("AC", 2000), ("Fridge", 1800)]

    def optimizePowerConsumption(device):
        return [device for device, power in device if power <= 1000]

    optimized_devices = optimizePowerConsumption(device)
    print("Active Devices:", optimized_devices)




#*****************SMART HOME AI ASSISTANT *****************





def smartHomeAIAssistant():
    def smartHomeAssistant(name, *args, **kwargs):
        print(f"Checking status for: {', '.join(args)}")
        for device in args:
            device_lower = device.lower()
            if device_lower in kwargs:
                print(f"{device} set to {kwargs[device_lower].upper()}")
            else:
                print(f"{device} has no status command.")

    smartHomeAssistant("Hiba", "Printer", "AC", printer="off", ac="on")




if __name__ == "__main__":
    print("WELCOME TO SMART HOME AUTOMATION SYSTEM")

    while True:
        print("\nMain Menu:")
        print("1. DEVICE MANAGEMENT OPTIONS")
        print("2. ENERGY CONSUMPTION TRACKER")
        print("3. SMART HOME AUTOMATION")
        print("4. SET AUTOMATION RULE")
        print("5. SMART HOME AI ASSISTANT")
        print("6. VIEW DEVICES WITH POWER")
        print("7. POWER SAVING MODES")
        print("8. EXIT")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            deviceManagement()
        elif choice == 2:
            energyConsumption()
        elif choice == 3:
            smartHomeAutomation()
        elif choice == 4:
            setAutomationRule()
        elif choice == 5:
            smartHomeAIAssistant()
        elif choice == 6:
            fan = ("fan", 'ON', '90W')
            lights = ("Lights", "ON", "60W")
            AC = ("AC", "OFF", "2000W")
            microwave = ("Microwave", "OFF", "1600W")
            doorLock = ("doorLock", "ON", "230W")
            AllDevices = [fan, lights, AC, microwave, doorLock]
            print("\nDevices with their power consumption:")
            for device in AllDevices:
                print(f"Device: {device[0]}, Status: {device[1]}, Power: {device[2]}")
        elif choice == 7:
            powerSaving = {"Eco mode", "Night mode"}
            print(f"\nPower Saving Modes: {powerSaving}")
            addNewmode = input("Do you want to add a new power saving mode? (yes/no): ").lower()
            if addNewmode == "yes":
                newMode = input("Name the mode: ")
                powerSaving.add(newMode)
                print(f"New mode added: {newMode}")
                print(f"Updated Modes: {powerSaving}")
        elif choice == 8:
            print("Exiting system.")
            break
        else:
            print("Invalid input. Try again.")
```
